```python
# 포스기, 키오스크 만들기
# 메뉴, 음료 선택, 사이즈, 타입(아이스, 핫), 토핑, 당도, 얼음양

import datetime
from typing import MutableSequence

menu = ['best_combination', 'original_tea', 'milk_tea', 'brown_sugar', 'fruit_mix', 'smoothie', 'coffee']

menu_dict = {
    'best_combination':  [
        {'drink':'블랙밀크티+펄', 'price':4200, 'size_up':1200, },
        {'drink':'타로밀크티+펄', 'price':4200, 'size_up':1200, },
        {'drink':'망고요구르트+화이트펄', 'price':5200, 'size_up':1200, },
        {'drink':'제주그린밀크티+펄', 'price':4900, 'size_up':1200, },
        {'drink':'패션후르츠히비스커스+코코넛', 'price':4500, 'size_up':1200, },
        {'drink':'우롱티+밀크폼', 'price':3700, 'size_up':1200, }
    ],  
    'original_tea': [
        {'drink':'자스민그린티', 'price':3500, 'size_up':1100, },
        {'drink':'우롱티', 'price':3500, 'size_up':1100, },
        {'drink':'블랙티', 'price':3500, 'size_up':1100, },
        {'drink':'얼그레이티', 'price':3500, 'size_up':1100, }
    ],
    'milk_tea': [
        {'drink':'민트초코칩밀크티+펄', 'price':4900, 'size_up':0, },
        {'drink':'복숭아쥬얼리밀크티', 'price':5100, 'size_up':0, },
        {'drink':'피스타치오밀크티+펄', 'price':4900, 'size_up':0, },
        {'drink':'망고치즈폼밀크티', 'price':4900, 'size_up':0, },
        {'drink':'딸기쥬얼리밀크티', 'price':5100, 'size_up':0, },
        {'drink':'자스민그린밀크티', 'price':4000, 'size_up':1100, },
        {'drink':'우롱밀크티', 'price':4000, 'size_up':1100, },
        {'drink':'블랙밀크티', 'price':4000, 'size_up':1100, },
        {'drink':'얼그레이밀크티', 'price':4000, 'size_up':1100, },
        {'drink':'타로밀크티', 'price':4000, 'size_up':1100, },
        {'drink':'초콜렛밀크티', 'price':4000, 'size_up':1100, },
        {'drink':'제주그린밀크티', 'price':4700, 'size_up':1100, },
        {'drink':'딸기쏙쏙쥬얼리밀크티', 'price':5100, 'size_up':0, },
        {'drink':'딸기쿠키밀크티', 'price':4900, 'size_up':0, }
    ], 
    'brown_sugar': [
        {'drink':'브라운슈가치즈폼쿠키밀크티', 'price':5700, 'size_up':0, },
        {'drink':'브라운슈가커피밀크티', 'price':5700, 'size_up':0, },
        {'drink':'브라운슈가쥬얼리밀크티', 'price':5300, 'size_up':0, },
        {'drink':'브라운슈가치즈폼스무디', 'price':5700, 'size_up':0, }
    ],
    'fruit_mix': [
        {'drink':'망고쥬스', 'price':4600, 'size_up':1100, },
        {'drink':'망고요구르트', 'price':5000, 'size_up':1100, },
        {'drink':'자몽그린티', 'price':4300, 'size_up':1100, },
        {'drink':'청포도그린티', 'price':4300, 'size_up':1100, },                
        {'drink':'패션후르츠히비스커스', 'price':4300, 'size_up':1100, },
        {'drink':'자몽주스', 'price':4600, 'size_up':1100, }
    ],
    'smoothie': [
        {'drink':'민트팝핑초코스무디', 'price':5300, 'size_up':0, },
        {'drink':'초코바른민트쿠키스무디', 'price':5300, 'size_up':0, },
        {'drink':'복숭아쥬얼리요구르트크러쉬', 'price':5300, 'size_up':0, },
        {'drink':'초코바른피스타치오스무디', 'price':5300, 'size_up':0, },
        {'drink':'망고요구르트쥬얼리크러쉬', 'price':5300, 'size_up':0, },
        {'drink':'돌체크러쉬위드샷', 'price':5300, 'size_up':0, },
        {'drink':'초코바른제주그린스무디', 'price':5700, 'size_up':0, },
        {'drink':'초코바른초코스무디', 'price':5500, 'size_up':0, },
        {'drink':'딸기&요구르트크러쉬', 'price':5300, 'size_up':0, },
        {'drink':'치즈폼딥초코스무디', 'price':5300, 'size_up':0, },
        {'drink':'망고스무디', 'price':5300, 'size_up':0, },
        {'drink':'초코쿠앤크스무디', 'price':4500, 'size_up':0, },
        {'drink':'청포도스무디', 'price':4800, 'size_up':0, },
        {'drink':'제주그린티스무디', 'price':5300, 'size_up':0, },
        {'drink':'체리바른딸기크러쉬', 'price':5500, 'size_up':0, },
        {'drink':'딸기치즈쿠키크러쉬', 'price':5500, 'size_up':0, },
        {'drink':'딸기쿠키스무디', 'price':5300, 'size_up':0, }
    ],
    'coffee': [
        {'drink':'돌체카페라떼', 'price':4700, 'size_up':1100, },
        {'drink':'카라멜카페라떼', 'price':4500, 'size_up':1100, },
        {'drink':'바닐라카페라떼', 'price':4500, 'size_up':1100, },
        {'drink':'모카카페라떼', 'price':4500, 'size_up':1100, },
        {'drink':'아메리카노', 'price':3500, 'size_up':500, },
        {'drink':'카페라떼', 'price':4000, 'size_up':1100, }
    ]
}

size = ['L', 'J']

sugar = ['0%', '30%', '50%', '70%', '100%']

type = ['ICE', 'HOT']

ice = ['없음', 'Less', 'Regular', 'Full']

topping = [
    {'name':'밀크폼', 'price':700, 'size_up':700},
    {'name':'펄', 'price':700, 'size_up':700},
    {'name':'코코넛', 'price':700, 'size_up':700},
    {'name':'알로에', 'price':700, 'size_up':700},
    {'name':'화이트펄', 'price':700, 'size_up':700}
]


def receipt(order):
    now = datetime.datetime.now()
    nowDatetime = now.strftime('%Y-%m-%d %H:%M:%S') 

    receipt_paper = '''
        공차 성수점
        대표: 조민서
        주소: 서울시 떡잎구 초록동
        전번: 0102-22933-1

        오늘도 공차를 찾아주셔서 감사합니다.
        '''
    print(receipt_paper)
    print('---------------------------------------------------------------------------')
    print('메뉴                                                                   금액')
    print('{0}{1}  당도{2}  타입{3}  얼음량{4}  {5}추가          {6}'.format(order['beverage'], order['size'], order['sugar'], order['type'], order['ice'], order['topping'], order['total_price']))
    print(sep='\n')
    print('---------------------------------------------------------------------------')
    print('합계 {0}원'.format(order['total_price']))
    print('판매일 ', nowDatetime)


def print_categories(menu):
    print('환영합니다. 공차입니다.\n카테고리를 골라주세요.\n')
    for idx, el in enumerate(menu):    # 'best_combination', 'original_tea', 'milk_tea', 'brown_sugar', 'fruit_mix', 'smoothie', 'coffee'
        print(idx + 1, '.', el)


def order_drink(menu):
    order = {}
    while True:
        input_category_choice = input('카테고리 -> 숫자를 입력하세요: ')
        # try에 오류 발생하면 except 수행한다.
        try:
            category_choice = int(input_category_choice)
            if category_choice != 0 and category_choice <= len(menu) :
                break
            continue
        except: 
            print('숫자로 입력해주세요.')
    selected_category = menu[category_choice-1]   # 'best_combination'

    # 음료 고르기
    for idx, el in enumerate(menu_dict[selected_category]):
        print(idx + 1, '.', el['drink'], '-', el['price'])
    while True:
        input_menu_choice = input('음료 -> 숫자를 입력하세요: ')
        try:
            menu_choice = int(input_menu_choice)
            if menu_choice != 0 and menu_choice <= len(menu_dict[selected_category]):
                break
            continue
        except:
            print('올바른 숫자를 입력해주세요.')

    selected_menu = menu_dict[selected_category][menu_choice-1]   # {'drink':'블랙밀크티+펄', 'price':4200, 'size_up':1200, }
    order['beverage'] = selected_menu['drink']   # 음료이름 저장
    print(order)

    # 사이즈 고르기
    for idx, el in enumerate(size):   # 'L', 'J'
        print(idx + 1, '.', el)
    while True:
        input_size_choice = input('사이즈 -> 숫자를 입력하세요: ')   # 1 or 2 or 3
        try:
            size_choice = int(input_size_choice)
            if size_choice != 0 and size_choice <= len(size):
                break
            continue
        except:
            print('올바른 숫자를 입력해주세요.')

    order['size'] = size[size_choice-1]   # 사이즈 저장
    print(order)
    total = selected_menu['price'] 
    if size_choice == 2:
        total += selected_menu['size_up']

    # 당도 고르기
    for idx, el in enumerate(sugar):   # 0%, 30%, 50%, 70%, 100%
        print(idx + 1, '.', el)
    while True:
        input_sugar_choice = input('당도 -> 숫자를 입력하세요: ') 
        try:
            sugar_choice = int(input_sugar_choice)
            if sugar_choice != 0 and sugar_choice <= len(sugar):
                break
            continue
        except:
            print('올바른 숫자를 입력해주세요.')
    order['sugar'] = sugar[sugar_choice-1]   # 당도 저장
    print(order)

    # 타입(ICE/HOT) 고르기 
    for idx, el in enumerate(type):
        print(idx + 1, '.', el)
    while True:
        input_type_choice = input('타입 -> 숫자를 입력하세요: ')
        try:
            type_choice = int(input_type_choice)
            if type_choice != 0 and type_choice <= len(type):
                break
            continue
        except:
            print('올바른 숫자를 입력해주세요.')
    order['type'] = type[type_choice-1]   # 타입 저장 
    print(order)  
    
    # 얼음량 고르기
    if type_choice == 1:
        for idx, el in enumerate(ice):
            print(idx + 1, '.', el)
        while True:
            input_ice_choice = input('얼음량 -> 숫자를 입력하세요: ')
            try:
                ice_choice = int(input_ice_choice)
                if ice_choice != 0 and ice_choice <= len(ice):
                    break
                continue
            except:
                print('올바른 숫자를 입력해주세요.')
        order['ice'] = ice[ice_choice-1]   # 얼음량 저장
        print(order)

    elif type_choice == 2:
        order['ice'] = ice[0]   # 얼음량 저장
        print(order)

    # 토핑 고르기
    topping_choice = order_topping(size_choice)
    total += topping[topping_choice-1]['price']
    if size_choice == 2:
        total += topping[topping_choice-1]['size_up']
    order['total_price'] = total   # total 저장
    order['topping'] = topping[topping_choice-1]['name']   # 토핑 저장

    print(order)
    print('총 ', total, '원입니다.')

    return order


# size에 영향받음
def order_topping(size_choice):
    for idx, el in enumerate(topping):
        if size_choice == 1:
            print(idx + 1, '.', el['name'], '-', el['price'])
        elif size_choice == 2:
            print(idx + 1, '.', el['name'], '-', el['price'] + el['size_up'])
    while True:
        input_topping_choice = input('토핑 -> 숫자를 입력하세요: ')
        try:
            topping_choice = int(input_topping_choice)
            if topping_choice != 0 and topping_choice <= len(topping):
                break
            continue
        except:
            print('올바른 숫자를 입력해주세요.')
    
    return topping_choice


def main():
    print_categories(menu)
    order = order_drink(menu)
    receipt(order)

if __name__ == '__main__':
    main()
```
